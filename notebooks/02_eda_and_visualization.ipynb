```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("../data/machine_logs.csv")
df=pd.DataFrame(data)

df_le=df.copy()
print(df_le.head())
print(df_le.shape)
```

```python
#Corelation
numeric_columns=['spindle_speed_rpm','vibration_mm_s','temperature_c','failure']

corr_matrix=df_le[numeric_columns].corr()

print("---Corelation Matrix---")
display(corr_matrix)
```

```python
# Visualizing sensor correlations to identify failure triggers
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Sensor Correlation Heatmap (Ti-6Al-4V Machining)')
```
