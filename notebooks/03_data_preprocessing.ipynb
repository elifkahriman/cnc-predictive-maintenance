```python
import pandas as pd
```

```python
data=pd.read_csv('../data/machine_logs.csv')
df=pd.DataFrame(data)
df_le=df.copy()
df_le.isnull().sum()
```

```python
mode_value=df_le['machine_id'].mode()[0]
df_le['machine_id']=df['machine_id'].fillna(mode_value)

df_le['vibration_mm_s']=df_le['vibration_mm_s'].interpolate(method='linear')
df_le['temperature_c']=df_le['temperature_c'].interpolate(method='linear')

df_le['machine_id']=df_le['machine_id'].str.strip().str.upper()
print(f"Remaining Missing Values:{df_le.isnull().sum().sum()}")

df_le.to_csv('../data/cleaned_machine_logs.csv', index=False)
```
